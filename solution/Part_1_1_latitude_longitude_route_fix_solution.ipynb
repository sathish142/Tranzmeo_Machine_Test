### Using OSRM API to find the nearest driving location

```python
# name of output csv file 
filename = "../output_files/latitude_longitude_corrected_details.csv"
# writing to csv file 
with open(filename, 'w') as csvfile:
    # creating a csv writer object 
    csvwriter = csv.writer(csvfile)       
    # writing the fields 
    csvwriter.writerow(['latitude', 'longitude']) 
    for latitude,longitude in zip(latitude_longitude['latitude'], latitude_longitude['longitude']):
      url = 'http://router.project-osrm.org/nearest/v1/driving/'+str(longitude)+','+str(latitude)+'?number=1'
      payload={}
      headers = {}
      response = requests.request("GET", url, headers=headers, data=payload)
      # print(response.text)
      response = json.loads(response.text)
      # print(response)
      #print(json.dumps(response, indent=4, sort_keys=True))
      if response["code"] == 'Ok':
        output = [response['waypoints'][0]["location"][1], response["waypoints"][0]["location"][0]]
        #print(output)
        # writing the data rows 
        csvwriter.writerow(output)
    
csvfile.close
```
