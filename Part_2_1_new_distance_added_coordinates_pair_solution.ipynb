### Import Packages

```python
import pandas as pd
import math
import csv
```

### Load input file

```python
latitude_longitude = pd.read_csv("../input_files/latitude_longitude_details.csv")
```

### Generate a set of points which are 25 meters to the left and right of the given latitude and longitude

```python
# name of output csv file 
filename = "../output_files/new_distance_added_coordinates.csv"
# writing to csv file 
with open(filename, 'w') as csvfile:
  # creating a csv writer object 
  csvwriter = csv.writer(csvfile)       
  # writing the fields 
  csvwriter.writerow(['latitude_longitude right distance added', 'latitude_longitude left distance added']) 
  for latitude,longitude in zip(latitude_longitude['latitude'], latitude_longitude['longitude']):
    #Earth radius in KM
    earth_radius = 6371
    meters_add = 25
    pi = math.pi
    cos = math.cos
    meter_per_degree = (1 / ((2 * pi / 360) * earth_radius)) / 1000
    new_longitude_right = longitude + (meters_add * meter_per_degree) / cos(latitude * (pi / 180))
    new_longitude_left = longitude - (meters_add * meter_per_degree) / cos(latitude * (pi / 180))
    output_file_list = [str(latitude)+' '+str(new_longitude_right), str(latitude)+' '+str(new_longitude_left)]
    print(output_file_list)
    # writing the data rows 
    csvwriter.writerow(output_file_list)
    
csvfile.close
```
